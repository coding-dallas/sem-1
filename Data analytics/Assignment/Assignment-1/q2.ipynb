```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sns

file=pd.read_csv(r'C:\Users\kesav\OneDrive\Desktop\cleaned_data.csv')
city=pd.DataFrame(file,
                         columns = ['City_name','is_ready_to_move']) 

#city[city['City_name'] == 'Chennai']

filtered=city.loc[(city['City_name']=='Chennai') | (city['City_name']=='Bangalore') | (city['City_name']=='Hyderabad')]
filtered_new=filtered.loc[(filtered['is_ready_to_move']==True)]

chennai=[]
bang=[]
hyd=[]

for i in filtered_new['City_name']:
    if i=='Chennai':
        chennai.append(1)
    if i=='Bangalore':
        bang.append(2)
    if i=='Hyderabad':
        hyd.append(3)

total=chennai+bang+hyd


data = {'Chennai':len(chennai), 'Bangalore':len(bang), 'Hyderabad':len(hyd)}
courses = list(data.keys())
values = list(data.values())

pt.xlabel('city')
pt.ylabel('no of houses in each city')
pt.title('Bar plot')
pt.bar(courses, values, color ='green',
        width = 0.4)
    
```
