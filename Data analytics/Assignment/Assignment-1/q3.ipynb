```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sns

file=pd.read_csv(r'C:\Users\kesav\OneDrive\Desktop\cleaned_data.csv')
def convert_float(x):
    return float(x.replace(',',''))
file['Price']=file['Price'].apply(convert_float)

def remove_string(x):
    return x.replace('sq ft','')

file['Size']=file['Size'].apply(remove_string)
file['Size']=file['Size'].apply(convert_float)


city=pd.DataFrame(file,
                       columns = ['Price','Size']) 

print(city.size)
sorted = city.sort_values(by='Size')
mean_size=round((city['Size'].mean()))


city = city.drop_duplicates()

index_to_be_deleted = city[ city['Size'] < mean_size ].index

city.drop(index_to_be_deleted,inplace=True)



pt.figure(0)
pt.xlabel('sq ft')
pt.ylabel('price in lakhs')
pt.title('all prize')
pt.scatter(city['Size'],city['Price'])
pt.figure(1)
fig = pt.gcf()
fig.set_size_inches(7,7)
pt.ylim(0,2000)
pt.xlabel('sq ft')
pt.ylabel('price in lakhs')
pt.title('low price')
pt.scatter(city['Size'],city['Price']/100000)

    
    
```
