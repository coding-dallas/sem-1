```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sns

file=pd.read_csv(r'C:\Users\kesav\OneDrive\Desktop\cleaned_data.csv')
city=pd.DataFrame(file,
                       columns = ['Locality_Name'])
hotspot=city['Locality_Name'].value_counts()

loc=pd.DataFrame(file,
                       columns = ['City_name'])
asset_count=loc['City_name'].value_counts()
print(asset_count)

filtered=hotspot.loc[(hotspot>1000)]
pt.figure(0)
fig = pt.gcf()
fig.set_size_inches(30,16)
pt.bar(filtered.keys(),filtered,width=0.4)

pt.figure(1)
fig = pt.gcf()
fig.set_size_inches(30,16)
pt.bar(asset_count.keys(),asset_count,width=0.4)
```
