```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sns

file=pd.read_csv(r'C:\Users\kesav\OneDrive\Desktop\cleaned_data.csv')
city=pd.DataFrame(file,
                       columns = ['City_name','Sub_urban_name']) 


Mumbai=city.loc[(city['City_name']=='Mumbai')]
Ahmedabad=city.loc[(city['City_name']=='Ahmedabad')]
hotspotm=Mumbai['Sub_urban_name'].value_counts()
hotspota=Ahmedabad['Sub_urban_name'].value_counts()

labelm=['Mira Road and Beyond','Andheri to Dahisar','Navi Mumbai','Thane','Beyond Thane','Central Mumbai','Mumbai Harbour','Mumbai South',
      'Worli to Andheri','Beyond Panvel','Other','Mumbai Mira Road and Beyond']

labela=['Ahmedabad West','Ahmedabad North','SG Highway','Ahmedabad East','Ahmedabad South','Ahmedabad South','Other']


fig = pt.gcf()
fig.set_size_inches(8,8)
pt.figure(0)
pt.title('Mumbai')
pt.pie(hotspotm,labels=labelm,autopct='%1.1f%%')
fig = pt.gcf()
fig.set_size_inches(8,8)
pt.figure(1)
pt.title('Ahmedabad')
pt.pie(hotspota,labels=labela,autopct='%1.1f%%')
 
```
