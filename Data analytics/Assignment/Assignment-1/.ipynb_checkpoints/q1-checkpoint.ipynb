```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
import seaborn as sns

file=pd.read_csv(r'C:\Users\kesav\OneDrive\Desktop\cleaned_data.csv')
def convert_float(x):
    return float(x.replace(',',''))
file['Price']=file['Price'].apply(convert_float)
price=file['Price']/10000

price=price.values.tolist()
price.sort()

range=round(len(price)/3)

#splitting the ranges
low_range=price[0:range]
med_range=price[(range):(range*2)]
high_range=price[(range*2):len(price)]
pt.figure(0)
pt.xlabel('range')
pt.ylabel('frequency')
pt.title('Histogram for high range')
pt.hist(high_range,ec='black')
high=[]
#removing outliers for high range
for i in high_range:
    if i<=2000:
        high.append(i)

total=[low_range,med_range,high]
pt.figure(1)
pt.xlabel('range')
pt.ylabel('value')
pt.title('Box plot')
pt.boxplot(total,labels=['low','mid','high'])
```
