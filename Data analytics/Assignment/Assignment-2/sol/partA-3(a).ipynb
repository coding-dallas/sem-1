```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('C:/Users/kesav/OneDrive/Desktop/Academics/Data analytics/Assignment/Assignment-2/data.csv')
```

```python
Active=data[((data['City_name']=='Hyderabad') | (data['City_name']=='Chennai') | 
           (data['City_name']=='Bangalore') | (data['City_name']=='Mumbai')) & 
           (data['Property_building_status']=='ACTIVE')]
```

```python
threshold=17
for i in data.columns:
    temp=data[i].drop_duplicates()
    if(len(temp)>threshold):
        Active.pop(i)
    elif data[i].dtype=='bool':
        Active.pop(i)
    elif len(data[i].drop_duplicates())==1:
        Active.pop(i)
```

```python
#Generalization based on no of distinct values
att_thre=4
for i in Active.columns:
    if len(Active[i].drop_duplicates())>=att_thre:
        if i == "No_of_BHK":
            Active['No_of_BHK'] = Active['No_of_BHK'].replace(['0 BHK','1 RK','1 BHK','2 RK','2 BHK','3 BHK','3 RK','4 BHK'],'Small house')
            Active['No_of_BHK'] = Active['No_of_BHK'].replace(['5 BHK','6 BHK','7 BHK','8 BHK','9 BHK'],'Luxury')
            Active['No_of_BHK'] = Active['No_of_BHK'].replace(['10 BHK','11 BHK','12 BHK','14 BHK'],'Bungalow')
        
        elif i=="Property_type":
            Active['Property_type']=Active['Property_type'].replace(['Apartment','Residential Plot'],'Shared')
            Active['Property_type']=Active['Property_type'].replace(['Villa','Independent House','Independent Floor'],'Individual')

#Droppping the redundant tuples
Active.pop('City_id')
newActive=Active.drop_duplicates()
newActive=newActive.drop_duplicates(['City_name','No_of_BHK','is_furnished'],keep='first')

```
