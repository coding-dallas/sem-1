```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('C:/Users/kesav/OneDrive/Desktop/Academics/Data analytics/Assignment/Assignment-2/data.csv')
# with pd.option_context("display.max_columns", None):
#     display(data.head())
```

```python
def convert_float(x):
    return float(x.replace(',',''))
data['Price_per_unit_area']=data['Price_per_unit_area'].apply(convert_float)
```

```python
data.loc[((data['Price_per_unit_area']>=100) & (data['Price_per_unit_area']<=10000)), 'Price_per_unit_area'] = 'Luxury'
```

```python
property=data[((data['City_name']=='Ahmedabad') | (data['City_name']=='Kolkata') | 
           (data['City_name']=='Delhi') | (data['City_name']=='Lucknow')) & 
           (data['Property_status']=='Under Construction') & 
           (data['Price_per_unit_area']=='Luxury')]
```

```python
threshold=17
for i in data.columns:
    temp=property[i].drop_duplicates()
    if(len(temp)>threshold):
        property.pop(i)
    elif data[i].dtype=='bool':
        property.pop(i)
    elif len(data[i].drop_duplicates())==1 and i!='Price_per_unit_area ':
        property.pop(i)
```

```python
#Generalization based on no of distinct values
att_thre=5
for i in property.columns:
    if len(property[i].drop_duplicates())>=att_thre:
        if i == "No_of_BHK":
            property['No_of_BHK'] = property['No_of_BHK'].replace(['0 BHK','1 RK','1 BHK','2 RK','2 BHK','3 BHK','3 RK','4 BHK'],'Small house')
            property['No_of_BHK'] = property['No_of_BHK'].replace(['5 BHK','6 BHK','7 BHK','8 BHK','9 BHK'],'Luxury')
            property['No_of_BHK'] = property['No_of_BHK'].replace(['10 BHK','11 BHK','12 BHK','14 BHK'],'Bungalow')
        
        elif i=="Property_type":
            property['Property_type']=property['Property_type'].replace(['Apartment','Residential Plot'],'Shared')
            property['Property_type']=property['Property_type'].replace(['Villa','Independent House','Independent Floor'],'Individual')
        else:
            property.pop(i)

```

```python
#Droppping the redundant & unwante tuples
property.pop('City_id')
newproperty=property.drop_duplicates()
newproperty=newproperty.drop_duplicates(['City_name','No_of_BHK','is_furnished'],keep='first')
```
