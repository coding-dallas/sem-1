```python
Tech=data[(data['City_name']=='Hyderabad') | (data['City_name']=='Chennai') | (data['City_name']=='Bangalore') | (data['City_name']=='Mumbai')]

NonTech=data[(data['City_name']=='Ahmedabad') | (data['City_name']=='Kolkata') | (data['City_name']=='Delhi') | (data['City_name']=='Lucknow')]

```

```python
threshold=17
for i in data.columns:
    temp=data[i].drop_duplicates()
    if(len(temp)>threshold):
        Tech.pop(i)
        NonTech.pop(i)
    elif data[i].dtype=='bool':
        Tech.pop(i)
        NonTech.pop(i)
    elif len(data[i].drop_duplicates())==1:
        Tech.pop(i)
        NonTech.pop(i)
```

```python
#Generalization based on no of distinct values
att_thre=5
for i in Tech.columns:
    if len(Tech[i].drop_duplicates())>att_thre:
        if i == "No_of_BHK":
            Tech['No_of_BHK'] = Tech['No_of_BHK'].replace(['0 BHK','1 RK','1 BHK','2 RK','2 BHK','3 BHK','3 RK','4 BHK'],'Small house')
            Tech['No_of_BHK'] = Tech['No_of_BHK'].replace(['5 BHK','6 BHK','7 BHK','8 BHK','9 BHK'],'Luxury')
            Tech['No_of_BHK'] = Tech['No_of_BHK'].replace(['10 BHK','11 BHK','12 BHK','14 BHK'],'Bungalow')
            
            NonTech['No_of_BHK'] = NonTech['No_of_BHK'].replace(['0 BHK','1 RK','1 BHK','2 RK','2 BHK','3 BHK','3 RK','4 BHK'],'Small house')
            NonTech['No_of_BHK'] = NonTech['No_of_BHK'].replace(['5 BHK','6 BHK','7 BHK','8 BHK','9 BHK'],'Luxury')
            NonTech['No_of_BHK'] = NonTech['No_of_BHK'].replace(['10 BHK','11 BHK','12 BHK','14 BHK'],'Bungalow')

        else:
            Tech.pop(i)
            NonTech.pop(i)

#Droppping the redundant tuples
newTech = Tech.drop_duplicates(
  subset = ['No_of_BHK', 'City_name']).reset_index(drop = True)

newNonTech = NonTech.drop_duplicates(
  subset = ['No_of_BHK', 'City_name']).reset_index(drop = True)
```
