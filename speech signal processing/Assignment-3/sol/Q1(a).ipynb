```python
from scipy.io import wavfile
import librosa
import numpy as np
from matplotlib import pyplot as plt
import statsmodels.api as sm
from scipy.signal import find_peaks
from statsmodels.graphics.tsaplots import plot_acf
import pandas as pd
```

```python
fileName='C:/Users/kesav/OneDrive/Desktop/Academics/speech signal processing/Assignment-3/o1.wav'
data,fs=librosa.load(fileName,sr=8000)
window = np.hamming(512)
```

```python
signal=data[0:len(window)]
windowing=signal*window

acorr=sm.tsa.acf(windowing,nlags=512)

plt.plot(acorr)
```
