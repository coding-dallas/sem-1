```python
from scipy.io import wavfile
import librosa
import numpy as np
from matplotlib import pyplot as plt
import statsmodels.api as sm
from scipy.fftpack import fft,fftfreq
from scipy.signal import find_peaks
```

```python
fileName='C:/Users/kesav/OneDrive/Desktop/Academics/speech signal processing/Assignment-3/o1.wav'
data,fs=librosa.load(fileName,sr=8000)
window = np.hamming(512)
```

```python
signal=data[0:len(window)]
windowing=signal*window

acorr=sm.tsa.acf(windowing,nlags=9000)
```

```python
peaks = find_peaks(acorr)[0]
lag = peaks[1]-peaks[0]
pitch = fs / lag
print(pitch)
```

```python
FFT = fft(windowing)
magnitude=abs(FFT)

n = len(magnitude)
fstep=fs/n
freq = np.linspace(0,(n-1)*fstep,n) 
fplot=freq[0:int(n/2)]

plt.plot(freq,magnitude)

peaks = find_peaks(magnitude)[0]
magnitude=magnitude.tolist()
pitch_index=magnitude.index(max(magnitude))
print(freq[peaks[2]])
```
