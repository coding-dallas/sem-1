```python
import numpy as np
import math
import scipy
from scipy.io import wavfile
from matplotlib import pyplot as plt

sample_rate, signal = wavfile.read('C:/Users/kesav/OneDrive/Desktop/Academics/speech signal processing/Assignment-2/H_MKB.wav')

frame_size = 0.025
frame_stride = 0.01


frame_length,frame_step = frame_size * sample_rate, frame_stride * sample_rate
frame_length=round(frame_length)

signal_length=len(signal)
num_frames =round(signal_length/frame_length)

frame_index=[0]
temp=0

for i in range(1,num_frames):
    temp+=frame_length
    frame_index.append(temp)
    

frame_no=23

x=range(frame_index[frame_no],frame_index[frame_no]+frame_length)

plt.figure(0)
plt.plot(x,signal[frame_index[frame_no]:frame_index[frame_no]+frame_length])
plt.xlabel('samples')
plt.ylabel('amplitude')
    
voiced=signal[frame_index[frame_no]:frame_index[frame_no]+frame_length]

#plot the FFT
fft = np.fft.fft(voiced)
t=range(frame_length)

plt.figure(1)
plt.plot(t,fft)
plt.title('FFT')
plt.xlabel('frequency')
plt.ylabel('amplitude')

#plot the log
ylog = np.log(fft)

plt.figure(2)
plt.plot(t,ylog)

#plot the exponential
exp=[]
for i in ylog:
    exp.append(math.exp(i))

#plot the inver fourier
ifft= np.fft.ifft(fft)
plt.figure(3)
plt.plot(t,ifft)
plt.xlabel('time')
plt.ylabel('amp')
plt.title('inverse FT')
```
