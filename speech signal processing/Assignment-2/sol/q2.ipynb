```python
from scipy.io import wavfile
import numpy as np
from matplotlib import pyplot as plt
from scipy.signal import correlate, find_peaks
import statsmodels.api as sm
from statsmodels.graphics import tsaplots
```

```python
fileName='C:/Users/kesav/OneDrive/Desktop/Academics/speech signal processing/Assignment-2/lataji_nrm.wav'
samplerate,amp=wavfile.read(fileName)
length = amp.shape[0]
t = length / samplerate
```

```python
time = np.linspace(0, length, amp.shape[0])
plt.plot(time,amp)
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('Time domain plot')
```

```python
frame_size = 0.025
frame_stride = 0.01

frame_length,frame_step = frame_size * samplerate, frame_stride * samplerate
frame_length=round(frame_length)
print(frame_length)

signal_length=len(amp)
num_frames =round(signal_length/frame_length)

frame_index=[0]
temp=0

for i in range(1,num_frames):
    temp+=frame_length
    frame_index.append(temp)
        

frame_no=3


x=range(frame_index[frame_no],frame_index[frame_no]+frame_length)
y=amp[frame_index[frame_no]:frame_index[frame_no]+frame_length]
# plt.figure(0)
# plt.plot(x,y)
```

```python
auto = sm.tsa.acf(y, nlags=0000)
plt.plot(auto)
```

```python
peaks = find_peaks(auto)[0]
print(peaks)
lag = peaks[0]
pitch = 1 / (peaks[1]-peaks[0])
print(pitch)
```

```python
pitch = samplerate / np.mean(np.diff(peaks))
print(pitch)
```
